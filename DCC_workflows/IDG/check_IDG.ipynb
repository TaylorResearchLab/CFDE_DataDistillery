```python
import csv
def get_delimiter(file_path: str) -> str:
    with open(file_path, 'r') as csvfile:
        delimiter = str(csv.Sniffer().sniff(csvfile.read()).delimiter)
        return delimiter
```

### change the SAB `PUBCHEM_CID` --> `PUBCHEM`

```python
# replace PUBCHEM_CID with just PUBCHEM
def replace_pubchem(column):
    return [i.replace('PUBCHEM_CID','PUBCHEM') if i.startswith('PUBCHEM_CID') \
                else i for i in column]
```

```python
nodes1 = pd.read_csv(PATH+'compound_disease/OWLNETS_node_metadata.txt',sep='\t')
nodes1['node_id'] = replace_pubchem(nodes1['node_id'])
nodes1.head(2)
```

```python
nodes1.to_csv(PATH+'compound_disease/OWLNETS_node_metadata.txt',sep='\t',index=False)
```

```python
nodes2 = pd.read_csv(PATH+'compound_protein/OWLNETS_node_metadata.txt',sep='\t')
nodes2['node_id'] = replace_pubchem(nodes2['node_id'])
nodes2.head(2)
```

```python
nodes2.to_csv(PATH+'compound_protein/OWLNETS_node_metadata.txt',sep='\t',index=False)
```
